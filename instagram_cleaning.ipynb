```python
import pandas as pd

df = pd.read_csv("instagram.csv")   # replace file name
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Null values in each column:")
print(df.isna().sum())
```

```python
# Fill numeric NaN with 0
df = df.fillna({
    "likes": 0,
    "comments": 0,
    "shares": 0,
    "saves": 0,
    "impressions": 0,
    "reach": 0,
    "views": 0,
    "followers_gained": 0
})

# Fill empty text columns
df = df.fillna("")
df
```

```python
df["upload_date"].unique()[:20]
```

```python
df = df.rename(columns={"upload_date": "duration"})
```

```python
df["duration"] = df["duration"].str.replace(".0", "", regex=False)
```

```python
df["duration_seconds"] = (
    df["duration"].str.split(":", expand=True)[0].astype(int) * 60 +
    df["duration"].str.split(":", expand=True)[1].astype(int)
)

df[["duration", "duration_seconds"]].head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
(df["impressions"] == 0).sum()
```

```python
(df["media_type"].str.strip() == "").sum()
```

```python
df = df[df["impressions"] > 0]
df = df[df["media_type"].str.strip() != ""]
```

```python
df["media_type"] = df["media_type"].str.strip().str.title()
```

```python
df.columns
```

```python
df["engagement_total"] = (
    df["likes"] + df["comments"] + df["shares"] + df["saves"]
)
df.head()
```

```python
df["engagement_rate"] = (
    (df["engagement_total"] / df["impressions"]) * 100
)
df.head()
```

```python
df = df.rename(columns={
    "upload_date": "date",
    "followers_gained": "followers"
})
```

```python
num_cols = ["likes", "comments", "shares", "saves", "impressions",
            "reach", "views", "followers"]

for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

df.head()
```

```python
df.to_csv("instagram_cleaned.csv", index=False)
print("Cleaning Completed Successfully!")
```
